```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# allmodels = np.load('/home/cheng/BayHunter-master/CH018/results/data/c_models.npy')
allmodels = np.load('tutorial/results/data/c_models.npy')
# allmodels = allmodels[~np.isnan(allmodels)]
```

```python
allmodels
```

```python
model_last = allmodels[-1]
model_last = model_last[~np.isnan(model_last)]
segment = int(len(model_last)/2)
x_age, y_rsl = model_last[:segment], model_last[segment:]
x_age, y_rsl
```

```python
station = 'NJ_CC.csv'
sea_level = pd.read_csv(station)
x_obs_raw = 1950-np.array(sea_level['Age'])
y_obs_raw = np.array(sea_level['RSL'])
sort_idx = np.argsort(x_obs_raw)
x_obs = x_obs_raw[sort_idx]
y_obs = y_obs_raw[sort_idx]

y_pred = np.interp(x_obs, x_age, y_rsl)

x_age_full = np.zeros(segment+2)
y_rsl_full = np.zeros(segment+2)
x_age_full[0] = x_obs[0]
x_age_full[1:-1] = x_age
x_age_full[-1] = x_obs[-1]
y_rsl_full[0] = y_pred[0]
y_rsl_full[1:-1] = y_rsl
y_rsl_full[-1] = y_pred[-1]
```

```python
plt.plot(x_obs, y_obs)
plt.plot(x_obs, y_pred)
```

```python
plt.plot(x_age_full[1:], (y_rsl_full[1:]-y_rsl_full[:-1])/(x_age_full[1:]-x_age_full[:-1])*100)
```

```python
plt.plot(x_obs, y_obs-y_pred, 'o')
```
